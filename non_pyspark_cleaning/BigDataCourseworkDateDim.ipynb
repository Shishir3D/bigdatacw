```python

calculated_total = (df_ord_itm['quantity'] * df_ord_itm['unit_price_at_sale']).round(2)
mismatch_mask = calculated_total != df_ord_itm['line_total'].round(2)
mismatched_rows = df_ord_itm[mismatch_mask].copy()

mismatched_rows['expected_total'] = calculated_total
mismatched_rows['difference'] = mismatched_rows['line_total'] - mismatched_rows['expected_total']

print(f"Found {len(mismatched_rows)} rows with calculation errors:")
mismatched_rows.head()
```
